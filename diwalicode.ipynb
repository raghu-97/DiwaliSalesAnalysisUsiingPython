```python
# importing library 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# import csv fileA
df = pd.read_csv(r'C:\Users\lenovo\OneDrive\Desktop\pythonprojectE2E\Python_Diwali_Sales_Analysis\Python_Diwali_Sales_Analysis\Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.drop(['Status', 'unnamed1'],axis=1, inplace=True )  
#Inplace assignment operations are widely popular in transforming Pandas DataFrames. As the name suggests, the core idea behind inplace assignment is to avoid creating a new DataFrame object 
#with each successive modification but instead
#making changes to the original DataFrame itsel
#inplace=True returns None inplace=False returns a copy of the object with the operation performed. 
```

```python
df.dropna(inplace=True)  #na to drop null value
```

```python

# change data type
df['Amount'] = df['Amount'].astype('int')  #column name always inside square bracket  #astype to change datatype
```

```python
df.columns
```

```python
#rename column
df.rename(columns= {'Marital_Status':'Marital_stat'},inplace=True)
```

```python
df['Marital_stat'] = df['Marital_stat'].replace({1: 'married', 0: 'non married'})
```

```python
df.rename(columns= {'Marital_Status':'Marital_stat'})
```

```python
ax=sns.countplot(x='Gender', data=df)
```

```python
ax=sns.countplot(x='Gender', data=df)  #for bar graph 
for bars in ax.containers:
 ax.bar_label(bars)  #for labeling bars
```

```python
colors = ["Blue", "Orange"]  
ax=sns.countplot(x='Gender', data=df, palette=colors)  #for bar graph 
for bars in ax.containers:
 ax.bar_label(bars)  #for labeling bars     # changing color using palette
```

```python
ax=sns.countplot(x='Gender', data=df, hue='Gender')  #for bar graph 
for bars in ax.containers:
 ax.bar_label(bars)  #for labeling bars
```

```python
AmountSpendByGender=df.groupby(['Gender'], as_index=False)['Gender'].count()
#print(AmountSpendByGender)
SumOfAmountSpendByGender=AmountSpendByGender['Amount'].sum()
SumOfAmountSpendByGender.sort_values(by='Gender', ascending=False)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, hue='Gender')
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python

# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age, hue='Age Group')
```
